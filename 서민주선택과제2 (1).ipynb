```python
f = open("실.txt", "r")
sil = f.read()
print(sil)
f.close
```

```python
tokens1 = nltk.wordpunct_tokenize(sil)
print(tokens1)
print(len(tokens1))
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('french')

print(stop_words)
```

```python
stop_words_capitalize = []

for i in stop_words:
  stop_words_capitalize.append(i.capitalize())

print(stop_words_capitalize)
```

```python

tokenized_doc = [i for i in tokens1 if i not in stop_words]
tokenized_doc = [i for i in tokenized_doc if i not in stop_words_capitalize]

print(tokenized_doc)
```

```python
stop_words_created = [',','-', '.', '(', ')', ';', '“', '.”','/', '’']
tokenized_doc = [i for i in tokenized_doc if i not in stop_words_created]

print(tokenized_doc)
```

```python
tagged = nltk.pos_tag(tokenized_doc)
tagged_new = dict(tagged)
print(tagged_new)
```

```python
text1 = " ".join(tokens1)
doc = nlp(text1)
data_new = " ".join([tokens1.lemma_ for tokens1 in doc])
print(data_new)
```

```python
import pickle

file = open('sil.pkl', 'wb')
pickle.dump(data_new, file)
file.close()
```
